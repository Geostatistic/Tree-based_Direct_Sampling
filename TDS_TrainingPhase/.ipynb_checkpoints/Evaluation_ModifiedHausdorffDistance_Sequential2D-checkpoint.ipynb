```python
# create the binary image
Binary_TrainingImage = np.zeros((Layer,Height,Width))
Binary_TrainingImage[ TrainingImage==0 ] = 1
Binary_TrainingImage[ TrainingImage==1 ] = 1

Binary_Realization_DS = np.zeros((RealizationAmount,Layer,Height,Width))
Binary_Realization_DS[ Realization_DS==0 ] = 1
Binary_Realization_DS[ Realization_DS==1 ] = 1

Binary_Realization_TDS = np.zeros((RealizationAmount,Layer,Height,Width))
Binary_Realization_TDS[ Realization_TDS==0 ] = 1
Binary_Realization_TDS[ Realization_TDS==1 ] = 1
```

```python
def ModifiedHausdorffDistance2D(image0,image1):
    threshold = 0.7
    
    edge_roberts0 = roberts(image0)
    edgeImage0 = np.copy(edge_roberts0)
    edgeImage0[edge_roberts0>threshold] = 1
    edgeImage0[edge_roberts0<=threshold] = 0
    edge_roberts1 = roberts(image1)
    edgeImage1 = np.copy(edge_roberts1)
    edgeImage1[edge_roberts1>threshold] = 1
    edgeImage1[edge_roberts1<=threshold] = 0
    
#     plt.figure(figsize=(5.4,5.4))
#     plt.imshow(edgeImage0,cmap='jet')
#     plt.title('The boundary in the '+str(point_z)+'-th layer of the training image')
# #     plt.colorbar(fraction=0.02)
#     plt.show()
    
#     plt.figure(figsize=(5.4,5.4))
#     plt.imshow(edgeImage1,cmap='jet')
#     plt.title('The boundary in the '+str(point_z)+'-th layer of the TDS realization')
# #     plt.colorbar(fraction=0.02)
#     plt.show()
    
    a = directed_hausdorff(edgeImage0, edgeImage1)[0]
    b = directed_hausdorff(edgeImage1, edgeImage0)[0]
#     print('the modified hausdorff distance is '+str(a)+' '+str(b))
    a = max(a,b)
    
    return a
```

```python
# test the modified hausdorff distance function
testLayer = 16
# plt.imshow(Binary_TrainingImage[testLayer,:,:],vmin=-1,vmax=2)
# plt.colorbar(fraction=0.02)
# plt.show()
# plt.imshow(Binary_Realization_DS[0,testLayer,:,:],vmin=-1,vmax=2)
# plt.colorbar(fraction=0.02)
# plt.show()
# plt.imshow(Binary_Realization_TDS[0,testLayer,:,:],vmin=-1,vmax=2)
# plt.colorbar(fraction=0.02)
# plt.show()
print('distance: '+str(ModifiedHausdorffDistance2D(Binary_TrainingImage[testLayer],Binary_Realization_DS[0,testLayer])))
print('distance: '+str(ModifiedHausdorffDistance2D(Binary_TrainingImage[testLayer],Binary_Realization_TDS[0,testLayer])))
```

```python
# define a distance matrix
MhDistanceMatrix = np.zeros((2*RealizationAmount+1,2*RealizationAmount+1))
```

```python
# compute the Hausdorff distance one by one DS realization vs Training image
averageMhDistance = 0.0
for realizationIndex in range(0, RealizationAmount):
    mhDistance_Z = 0.0
    for testLayer in range(0,Layer):
        mhDistance_Z += ModifiedHausdorffDistance2D(Binary_TrainingImage[testLayer],Binary_Realization_DS[realizationIndex,testLayer])
        # print(mhDistance)
    mhDistance_Z = mhDistance_Z / Layer
    
    mhDistance_Y = 0.0
    for testHeight in range(0,Height):
        mhDistance_Y += ModifiedHausdorffDistance2D(Binary_TrainingImage[:,testHeight,:],Binary_Realization_DS[realizationIndex,:,testHeight,:])
        # print(mhDistance)
    mhDistance_Y = mhDistance_Y / Height
    
    mhDistance_X = 0.0
    for testWidth in range(0,Width):
        mhDistance_X += ModifiedHausdorffDistance2D(Binary_TrainingImage[:,:,testWidth],Binary_Realization_DS[realizationIndex,:,:,testWidth])
        # print(mhDistance)
    mhDistance_X = mhDistance_X / Width   
    
    mhDistance = (mhDistance_X+mhDistance_Y+mhDistance_Z)/3.0
    
    MhDistanceMatrix[0][realizationIndex+1] = mhDistance
    MhDistanceMatrix[realizationIndex+1][0] = mhDistance
    averageMhDistance += mhDistance
    print('the mh Distance is '+str(mhDistance))
print('the average mh distance is '+str(averageMhDistance/RealizationAmount))
```

```python
# compute the Hausdorff distance one by one TDS realization vs Training image
averageMhDistance = 0.0
for realizationIndex in range(0, RealizationAmount):
    mhDistance_Z = 0.0
    for testLayer in range(0,Layer):
        mhDistance_Z += ModifiedHausdorffDistance2D(Binary_TrainingImage[testLayer],Binary_Realization_TDS[realizationIndex,testLayer])
        # print(mhDistance)
    mhDistance_Z = mhDistance_Z / Layer
    
    mhDistance_Y = 0.0
    for testHeight in range(0,Height):
        mhDistance_Y += ModifiedHausdorffDistance2D(Binary_TrainingImage[:,testHeight,:],Binary_Realization_TDS[realizationIndex,:,testHeight,:])
        # print(mhDistance)
    mhDistance_Y = mhDistance_Y / Height
    
    mhDistance_X = 0.0
    for testWidth in range(0,Width):
        mhDistance_X += ModifiedHausdorffDistance2D(Binary_TrainingImage[:,:,testWidth],Binary_Realization_TDS[realizationIndex,:,:,testWidth])
        # print(mhDistance)
    mhDistance_X = mhDistance_X / Width   
    
    mhDistance = (mhDistance_X+mhDistance_Y+mhDistance_Z)/3.0
    
    MhDistanceMatrix[0][realizationIndex+RealizationAmount+1] = mhDistance
    MhDistanceMatrix[realizationIndex+RealizationAmount+1][0] = mhDistance
    averageMhDistance += mhDistance
    print('the mh Distance is '+str(mhDistance))
print('the average mh distance is '+str(averageMhDistance/RealizationAmount))
```

```python
# among DS realizations
averageMhDistance = 0.0
for realizationIndex0 in range(0, RealizationAmount):
    for realizationIndex1 in range(0, RealizationAmount):
        mhDistance_Z = 0.0
        for testLayer in range(0,Layer):
            mhDistance_Z += ModifiedHausdorffDistance2D(Binary_Realization_DS[realizationIndex0,testLayer],Binary_Realization_DS[realizationIndex1,testLayer])
            # print(mhDistance)
        mhDistance_Z = mhDistance_Z / Layer

        mhDistance_Y = 0.0
        for testHeight in range(0,Height):
            mhDistance_Y += ModifiedHausdorffDistance2D(Binary_Realization_DS[realizationIndex0,:,testHeight,:],Binary_Realization_DS[realizationIndex1,:,testHeight,:])
            # print(mhDistance)
        mhDistance_Y = mhDistance_Y / Height

        mhDistance_X = 0.0
        for testWidth in range(0,Width):
            mhDistance_X += ModifiedHausdorffDistance2D(Binary_Realization_DS[realizationIndex0,:,:,testWidth],Binary_Realization_DS[realizationIndex1,:,:,testWidth])
            # print(mhDistance)
        mhDistance_X = mhDistance_X / Width   

        mhDistance = (mhDistance_X+mhDistance_Y+mhDistance_Z)/3.0

        MhDistanceMatrix[realizationIndex0+1][realizationIndex1+1] = mhDistance
        MhDistanceMatrix[realizationIndex1+1][realizationIndex0+1] = mhDistance
        averageMhDistance += mhDistance
        print('the mh Distance is '+str(mhDistance))
print('the average mh distance is '+str(averageMhDistance/(RealizationAmount*RealizationAmount)))
```

```python
# TDS versus TDS
averageMhDistance = 0.0
for realizationIndex0 in range(0, RealizationAmount):
    for realizationIndex1 in range(0, RealizationAmount):
        mhDistance_Z = 0.0
        for testLayer in range(0,Layer):
            mhDistance_Z += ModifiedHausdorffDistance2D(Binary_Realization_TDS[realizationIndex0,testLayer],Binary_Realization_TDS[realizationIndex1,testLayer])
            # print(mhDistance)
        mhDistance_Z = mhDistance_Z / Layer

        mhDistance_Y = 0.0
        for testHeight in range(0,Height):
            mhDistance_Y += ModifiedHausdorffDistance2D(Binary_Realization_TDS[realizationIndex0,:,testHeight,:],Binary_Realization_TDS[realizationIndex1,:,testHeight,:])
            # print(mhDistance)
        mhDistance_Y = mhDistance_Y / Height

        mhDistance_X = 0.0
        for testWidth in range(0,Width):
            mhDistance_X += ModifiedHausdorffDistance2D(Binary_Realization_TDS[realizationIndex0,:,:,testWidth],Binary_Realization_TDS[realizationIndex1,:,:,testWidth])
            # print(mhDistance)
        mhDistance_X = mhDistance_X / Width   

        mhDistance = (mhDistance_X+mhDistance_Y+mhDistance_Z)/3.0

        MhDistanceMatrix[realizationIndex0+1+RealizationAmount][realizationIndex1+1+RealizationAmount] = mhDistance
        MhDistanceMatrix[realizationIndex1+1+RealizationAmount][realizationIndex0+1+RealizationAmount] = mhDistance
        averageMhDistance += mhDistance
        print('the mh Distance is '+str(mhDistance))
print('the average mh distance is '+str(averageMhDistance/(RealizationAmount*RealizationAmount)))
```

```python
# DS versus TDS
averageMhDistance = 0.0
for realizationIndex0 in range(0, RealizationAmount):
    for realizationIndex1 in range(0, RealizationAmount):
        mhDistance_Z = 0.0
        for testLayer in range(0,Layer):
            mhDistance_Z += ModifiedHausdorffDistance2D(Binary_Realization_DS[realizationIndex0,testLayer],Binary_Realization_TDS[realizationIndex1,testLayer])
            # print(mhDistance)
        mhDistance_Z = mhDistance_Z / Layer

        mhDistance_Y = 0.0
        for testHeight in range(0,Height):
            mhDistance_Y += ModifiedHausdorffDistance2D(Binary_Realization_DS[realizationIndex0,:,testHeight,:],Binary_Realization_TDS[realizationIndex1,:,testHeight,:])
            # print(mhDistance)
        mhDistance_Y = mhDistance_Y / Height

        mhDistance_X = 0.0
        for testWidth in range(0,Width):
            mhDistance_X += ModifiedHausdorffDistance2D(Binary_Realization_DS[realizationIndex0,:,:,testWidth],Binary_Realization_TDS[realizationIndex1,:,:,testWidth])
            # print(mhDistance)
        mhDistance_X = mhDistance_X / Width   

        mhDistance = (mhDistance_X+mhDistance_Y+mhDistance_Z)/3.0

        MhDistanceMatrix[realizationIndex0+1][realizationIndex1+1+RealizationAmount] = mhDistance
        MhDistanceMatrix[realizationIndex1+1+RealizationAmount][realizationIndex0+1] = mhDistance
        averageMhDistance += mhDistance
        print('the mh Distance is '+str(mhDistance))
print('the average mh distance is '+str(averageMhDistance/(RealizationAmount*RealizationAmount)))
```

```python
# MDS display
from sklearn import manifold
mds = manifold.MDS(n_components=2, max_iter=300, eps=1e-9, dissimilarity="precomputed", n_jobs=1)
pos_Global = mds.fit(MhDistanceMatrix).embedding_
plt.scatter(pos_Global[1:RealizationAmount+1, 0], pos_Global[1:RealizationAmount+1, 1], color='forestgreen', s=30, lw=0, label='DS')
plt.scatter(pos_Global[RealizationAmount+1:2*RealizationAmount+1, 0], pos_Global[RealizationAmount+1:2*RealizationAmount+1, 1], color='deepskyblue', s=30, lw=0, label='TDS')
plt.scatter(pos_Global[0, 0], pos_Global[0, 1], color='red', s=70, lw=0, label='TI')
print('the size of pos_DS is '+str(pos_Global.shape))
plt.legend(scatterpoints=1, loc='best', shadow=False)
plt.xlabel(str('MDS1'))
plt.ylabel(str('MDS2'))
plt.show()
```

```python
# modified the matrix  random assign DS and TDS realizations
import random
list = []
for index in range(0,2*RealizationAmount+1):
    list.append(index)
print(list)
randomIndex = random.sample(list,k=2*RealizationAmount+1)
print(randomIndex)
randomIndex = np.array(randomIndex)
print(randomIndex)

label = np.zeros(2*RealizationAmount+1)
label[ randomIndex > 0] = 1
label[ randomIndex > RealizationAmount ] = 2
print(label)

DistanceMatrix = np.empty((2*RealizationAmount+1,2*RealizationAmount+1))

for i in range(0,2*RealizationAmount+1):
    for j in range(0,2*RealizationAmount+1):
        x = randomIndex[i]
        y = randomIndex[j]
        distance = MhDistanceMatrix[x][y]
        DistanceMatrix[i][j] = distance
        
mds = manifold.MDS(n_components=2, max_iter=3000, eps=1e-9, dissimilarity="precomputed", n_jobs=10)
pos_Global = mds.fit(DistanceMatrix).embedding_

plt.scatter(pos_Global[label==1, 0], pos_Global[label==1, 1], color='forestgreen', s=30, lw=0, label='DS')
plt.scatter(pos_Global[label==2, 0], pos_Global[label==2, 1], color='deepskyblue', s=30, lw=0, label='TDS')
plt.scatter(pos_Global[label==0, 0], pos_Global[label==0, 1], color='red', s=70, lw=0, label='TI')
print('the size of pos_DS is '+str(pos_Global.shape))

plt.legend(scatterpoints=1, loc='upper right', shadow=False,fontsize='large')
plt.xlabel(str('MDS1'),fontsize='large')
plt.ylabel(str('MDS2'),fontsize='large')
plt.tick_params(direction='in',labelsize='large')
plt.xlim(-5,7)
plt.ylim(-5,7)
plt.xticks(np.linspace(start=-6,stop=8,num=8))
plt.yticks(np.linspace(start=-6,stop=8,num=8))
plt.show()
```
