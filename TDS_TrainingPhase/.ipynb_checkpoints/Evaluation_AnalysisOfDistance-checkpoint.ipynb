```python
import matplotlib.pyplot as plt
import numpy as np
import scipy

from PIL import Image
from sklearn.cluster import KMeans
from sklearn import manifold
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
# set several parameters

RealizationAmountPerSet = 20   # the number of realizations for each algorithm

ScaleIndex_Max = 2   # the number of scales

template_radius_x = 2  # the size of template is 2*radius+1
template_radius_y = 2

ClusterAmountGroup = [30,20,10]   # the number of clusters in each scale
```

```python
# set some temporary variable
point_y = 0
point_x = 0
point_value = 0

sample_y = 0
sample_x = 0
sample_value = 0
realizationlist = []
```

```python
TIBMP = Image.open("ChannelRealizations\channel101.bmp",'r')

TIInt = TIBMP.load()
TI_Width_Orginal = TIBMP.size[0]
TI_Height_Orginal = TIBMP.size[1]
TI_Orginal = np.empty((TI_Height_Orginal,TI_Width_Orginal))

for point_y in range(TI_Height_Orginal):
    for point_x in range(TI_Width_Orginal):
        point_value = float(TIInt[point_x,point_y][0])
        TI_Orginal[point_y][point_x] = point_value/255;

print(TI_Width_Orginal)
print(TI_Height_Orginal)
# the horizontal direction is x-axis and corresponds to the width
# the vertical direction is y-axis and correspongds to the height
print("the training image is: ")
plt.imshow(TI_Orginal)
plt.colorbar();  # show color scale
plt.show();
```

```python
# read the DS realization
Realization_DS_Orginal = np.empty((RealizationAmountPerSet,TI_Height_Orginal,TI_Width_Orginal))
index_image = 0;
for index_image in range(0,RealizationAmountPerSet):
    TIBMP = Image.open("ChannelRealizations\Realization_DS"+str(index_image)+".bmp",'r')
    TIInt = TIBMP.load()
    for point_y in range(TI_Height_Orginal):
        for point_x in range(TI_Width_Orginal):
            point_value = float(TIInt[point_x,point_y][0])
            Realization_DS_Orginal[index_image][point_y][point_x] = point_value/255;
print("the first realization is: ")
plt.imshow(Realization_DS_Orginal[0],vmin=0, vmax=1)
plt.colorbar();  # show color scale
plt.show();
```

```python
# read the CTDS realization
Realization_PointCTDS_Orginal = np.empty((RealizationAmountPerSet,TI_Height_Orginal,TI_Width_Orginal))
for index_image in range(0,RealizationAmountPerSet):
    TIBMP = Image.open("ChannelRealizations\Realization_PointCTS"+str(index_image)+".bmp",'r')
    TIInt = TIBMP.load()
    for point_y in range(TI_Height_Orginal):
        for point_x in range(TI_Width_Orginal):
            point_value = float(TIInt[point_x,point_y][0])
            Realization_PointCTDS_Orginal[index_image][point_y][point_x] = point_value/255;
print("the first realization is: ")
plt.imshow(Realization_PointCTDS_Orginal[0],vmin=0, vmax=1)
plt.colorbar();  # show color scale
plt.show();
```

```python
# read the pattern-based Pattern Diversity DS realization
Realization_PatternCTDS_Orginal = np.empty((RealizationAmountPerSet,TI_Height_Orginal,TI_Width_Orginal))
for index_image in range(0,RealizationAmountPerSet):
    TIBMP = Image.open("ChannelRealizations\Realization_PatternCTS"+str(index_image)+".bmp",'r')
    TIInt = TIBMP.load()
    for point_y in range(TI_Height_Orginal):
        for point_x in range(TI_Width_Orginal):
            point_value = float(TIInt[point_x,point_y][0])
            Realization_PatternCTDS_Orginal[index_image][point_y][point_x] = point_value/255;
print("the first realization is: ")
plt.imshow(Realization_PatternCTDS_Orginal[0],vmin=0, vmax=1)
plt.colorbar();  # show color scale
plt.show();
```

```python
# main program multi-resolution analysis

ClusterAmount_Max = max(ClusterAmountGroup)

ClusterDistribution_TI = np.zeros((ScaleIndex_Max+1,ClusterAmount_Max))
KLDivergenceMatrix_DS = np.zeros((ScaleIndex_Max+1,RealizationAmountPerSet+1,RealizationAmountPerSet+1))
ClusterDistribution_Realization_DS = np.zeros((ScaleIndex_Max+1,RealizationAmountPerSet,ClusterAmount_Max))
print('the size of distance matrix is '+str(KLDivergenceMatrix_DS.shape))
KLDivergenceMatrix_PointCTDS = np.zeros((ScaleIndex_Max+1,RealizationAmountPerSet+1,RealizationAmountPerSet+1))
ClusterDistribution_Realization_PointCTDS = np.zeros((ScaleIndex_Max+1,RealizationAmountPerSet,ClusterAmount_Max))
KLDivergenceMatrix_PatternCTDS = np.zeros((ScaleIndex_Max+1,RealizationAmountPerSet+1,RealizationAmountPerSet+1))
ClusterDistribution_Realization_PatternCTDS = np.zeros((ScaleIndex_Max+1,RealizationAmountPerSet,ClusterAmount_Max))

for scaleIndex in range(0,ScaleIndex_Max+1):
    scaleZoom = 2**scaleIndex
    print('index: ' +str(scaleIndex)+' zoom: '+str(scaleZoom))
    ClusterAmount = ClusterAmountGroup[scaleIndex]
    # carry the downsample
    TI = TI_Orginal[::scaleZoom,::scaleZoom]
    TI_Height = TI.shape[0]
    TI_Width = TI.shape[1]
    print('the shape of images: height: '+str(TI_Height)+' width: '+str(TI_Width))
#     print("the training image is: ")
#     plt.imshow(TI)
#     plt.colorbar()  # show color scale
#     plt.show()
    
    Realization_DS = Realization_DS_Orginal[:,::scaleZoom,::scaleZoom]
#     print('the size of DS realizations is '+str(Realization_DS.shape))
    Realization_PointCTDS = Realization_PointCTDS_Orginal[:,::scaleZoom,::scaleZoom]
    Realization_PatternCTDS = Realization_PatternCTDS_Orginal[:,::scaleZoom,::scaleZoom]

    # analyse the TI
    # extract all pattern
    PatternAmount = (TI_Height-2*template_radius_y)*(TI_Width-2*template_radius_x);
    PatternSize = (2*template_radius_y+1)*(2*template_radius_x+1)
    PatternBase = np.empty((PatternAmount,PatternSize))
    index_Base = 0
    for point_y in range(template_radius_y,TI_Height-template_radius_y):
        for point_x in range(template_radius_x,TI_Width-template_radius_x):
            index_Pattern = 0
            for sample_y in range(point_y-template_radius_y,point_y+template_radius_y+1):
                for sample_x in range(point_x-template_radius_x,point_x+template_radius_x+1):
                    value = TI[sample_y][sample_x]
                    PatternBase[index_Base][index_Pattern] = value
                    index_Pattern += 1
            index_Base += 1
#     print("the amount of patterns is "+str(len(PatternBase)))
#     print("the size of patterns is "+str(len(PatternBase[0])))
    # print(PatternBase)
    
    # conduct K means to cluster all training patterns
    kmeans = KMeans(n_clusters= ClusterAmount, random_state=0).fit(PatternBase);
    KmeansResult = kmeans.labels_
    # print(KmeansResult)
    KmeansCenter = kmeans.cluster_centers_
#     print('the shape of KmeansCenter is '+ str(KmeansCenter.shape))
    # print(KmeansCenter)
    KmeansResult2D = KmeansResult.reshape(TI_Height-2*template_radius_y,TI_Width-2*template_radius_x)
#     print("the clustered image is: ")
#     plt.imshow(KmeansResult2D)
#     plt.colorbar()  # show color scale
#     plt.show()

    # analyze the TI
    ClusterDistribution_Temporary, HistogramEdge = np.histogram(KmeansResult,bins=np.linspace(0,ClusterAmount,ClusterAmount+1))
    ClusterDistribution_TI[scaleIndex][0:ClusterAmount] = ClusterDistribution_Temporary
#     print('the edge of histogram is '+str(HistogramEdge))
    ClusterDistribution_TI[scaleIndex] = np.divide(ClusterDistribution_TI[scaleIndex],sum(ClusterDistribution_TI[scaleIndex]))
    ClusterDistribution_TI[scaleIndex] = ClusterDistribution_TI[scaleIndex] + 0.000000001
#     print('the pattern hisrogram of TI is '+str(ClusterDistribution_TI[scaleIndex]))
#     plt.bar(HistogramEdge[:-1], ClusterDistribution_TI[scaleIndex][0:ClusterAmount], width = 0.8)
#     plt.show()
    
    # analyze DS realization
    for index_image in range(0,RealizationAmountPerSet):
        PatternBase_Realization = np.empty((PatternAmount,PatternSize))
        index_Base = 0
        for point_y in range(template_radius_y,TI_Height-template_radius_y):
            for point_x in range(template_radius_x,TI_Width-template_radius_x):
                index_Pattern = 0;
                for sample_y in range(point_y-template_radius_y,point_y+template_radius_y+1):
                    for sample_x in range(point_x-template_radius_x,point_x+template_radius_x+1):
                        value = Realization_DS[index_image][sample_y][sample_x];
                        PatternBase_Realization[index_Base][index_Pattern] = value;
                        index_Pattern += 1;
                index_Base += 1;
        for PatternIndex in range(PatternAmount):     
            OnePattern = PatternBase_Realization[PatternIndex]
            OnePattern = OnePattern.reshape(1,PatternSize)
            OnePattern = np.subtract(KmeansCenter,OnePattern)
            OnePattern = np.linalg.norm(OnePattern,axis=1)
            OnePattern = OnePattern.reshape(ClusterAmount,1)
            ArgMinIdex = OnePattern.argmin()
            ClusterDistribution_Realization_DS[scaleIndex][index_image][ArgMinIdex] += 1.0
        ClusterDistribution_Realization_DS[scaleIndex][index_image] = np.divide(ClusterDistribution_Realization_DS[scaleIndex][index_image],PatternAmount)
        ClusterDistribution_Realization_DS[scaleIndex][index_image] += + 0.000000001
    for index_image in range(0,RealizationAmountPerSet):
        KLDivergence = 0.5 * (scipy.stats.entropy(ClusterDistribution_TI[scaleIndex],ClusterDistribution_Realization_DS[scaleIndex][index_image]) + scipy.stats.entropy(ClusterDistribution_Realization_DS[scaleIndex][index_image],ClusterDistribution_TI[scaleIndex]))
        KLDivergenceMatrix_DS[scaleIndex][0][index_image+1] = KLDivergence
        KLDivergenceMatrix_DS[scaleIndex][index_image+1][0] = KLDivergence
    for point_y in range(0,RealizationAmountPerSet):
        for point_x in range(point_y+1,RealizationAmountPerSet):
            KLDivergence = 0.5 * (scipy.stats.entropy(ClusterDistribution_Realization_DS[scaleIndex][point_y],ClusterDistribution_Realization_DS[scaleIndex][point_x]) + scipy.stats.entropy(ClusterDistribution_Realization_DS[scaleIndex][point_x],ClusterDistribution_Realization_DS[scaleIndex][point_y]))
            KLDivergenceMatrix_DS[scaleIndex][point_y+1][point_x+1] = KLDivergence
            KLDivergenceMatrix_DS[scaleIndex][point_x+1][point_y+1] = KLDivergence
    # print('the shape of distance matrix is '+str(KLDivergenceMatrix_DS[scaleIndex].shape))
    # print('the distance matrix is\n')
    # print(KLDivergenceMatrix_DS[scaleIndex])
    # print('the distance with TI is '+str(KLDivergenceMatrix_DS[scaleIndex][0]))
    print('the mean distance between DS realizations and TI is '+str(np.mean(KLDivergenceMatrix_DS[scaleIndex][0])))
    
    # analyze Point_CTDS realization
    for index_image in range(0,RealizationAmountPerSet):
        PatternBase_Realization = np.empty((PatternAmount,PatternSize))
        index_Base = 0
        for point_y in range(template_radius_y,TI_Height-template_radius_y):
            for point_x in range(template_radius_x,TI_Width-template_radius_x):
                index_Pattern = 0;
                for sample_y in range(point_y-template_radius_y,point_y+template_radius_y+1):
                    for sample_x in range(point_x-template_radius_x,point_x+template_radius_x+1):
                        value = Realization_PointCTDS[index_image][sample_y][sample_x];
                        PatternBase_Realization[index_Base][index_Pattern] = value;
                        index_Pattern += 1;
                index_Base += 1;
        for PatternIndex in range(PatternAmount):     
            OnePattern = PatternBase_Realization[PatternIndex]
            OnePattern = OnePattern.reshape(1,PatternSize)
            OnePattern = np.subtract(KmeansCenter,OnePattern)
            OnePattern = np.linalg.norm(OnePattern,axis=1)
            OnePattern = OnePattern.reshape(ClusterAmount,1)
            ArgMinIdex = OnePattern.argmin()
            ClusterDistribution_Realization_PointCTDS[scaleIndex][index_image][ArgMinIdex] += 1.0
        ClusterDistribution_Realization_PointCTDS[scaleIndex][index_image] = np.divide(ClusterDistribution_Realization_PointCTDS[scaleIndex][index_image],PatternAmount)
        ClusterDistribution_Realization_PointCTDS[scaleIndex][index_image] += + 0.000000001
    for index_image in range(0,RealizationAmountPerSet):
        KLDivergence = 0.5 * (scipy.stats.entropy(ClusterDistribution_TI[scaleIndex],ClusterDistribution_Realization_PointCTDS[scaleIndex][index_image]) + scipy.stats.entropy(ClusterDistribution_Realization_PointCTDS[scaleIndex][index_image],ClusterDistribution_TI[scaleIndex]))
        KLDivergenceMatrix_PointCTDS[scaleIndex][0][index_image+1] = KLDivergence
        KLDivergenceMatrix_PointCTDS[scaleIndex][index_image+1][0] = KLDivergence
    for point_y in range(0,RealizationAmountPerSet):
        for point_x in range(point_y+1,RealizationAmountPerSet):
            KLDivergence = 0.5 * (scipy.stats.entropy(ClusterDistribution_Realization_PointCTDS[scaleIndex][point_y],ClusterDistribution_Realization_PointCTDS[scaleIndex][point_x]) + scipy.stats.entropy(ClusterDistribution_Realization_PointCTDS[scaleIndex][point_x],ClusterDistribution_Realization_PointCTDS[scaleIndex][point_y]))
            KLDivergenceMatrix_PointCTDS[scaleIndex][point_y+1][point_x+1] = KLDivergence
            KLDivergenceMatrix_PointCTDS[scaleIndex][point_x+1][point_y+1] = KLDivergence
    # print('the shape of distance matrix is '+str(KLDivergenceMatrix_PointCTDS[scaleIndex].shape))
    # print('the distance with TI is '+str(KLDivergenceMatrix_PointCTDS[scaleIndex][0]))
    print('the mean distance between Point CTDS realizations and TI is '+str(np.mean(KLDivergenceMatrix_PointCTDS[scaleIndex][0])))

    # analyze Pattern_CTDS realization
    for index_image in range(0,RealizationAmountPerSet):
        PatternBase_Realization = np.empty((PatternAmount,PatternSize))
        index_Base = 0
        for point_y in range(template_radius_y,TI_Height-template_radius_y):
            for point_x in range(template_radius_x,TI_Width-template_radius_x):
                index_Pattern = 0;
                for sample_y in range(point_y-template_radius_y,point_y+template_radius_y+1):
                    for sample_x in range(point_x-template_radius_x,point_x+template_radius_x+1):
                        value = Realization_PatternCTDS[index_image][sample_y][sample_x];
                        PatternBase_Realization[index_Base][index_Pattern] = value;
                        index_Pattern += 1;
                index_Base += 1;
        for PatternIndex in range(PatternAmount):     
            OnePattern = PatternBase_Realization[PatternIndex]
            OnePattern = OnePattern.reshape(1,PatternSize)
            OnePattern = np.subtract(KmeansCenter,OnePattern)
            OnePattern = np.linalg.norm(OnePattern,axis=1)
            OnePattern = OnePattern.reshape(ClusterAmount,1)
            ArgMinIdex = OnePattern.argmin()
            ClusterDistribution_Realization_PatternCTDS[scaleIndex][index_image][ArgMinIdex] += 1.0
        ClusterDistribution_Realization_PatternCTDS[scaleIndex][index_image] = np.divide(ClusterDistribution_Realization_PatternCTDS[scaleIndex][index_image],PatternAmount)
        ClusterDistribution_Realization_PatternCTDS[scaleIndex][index_image] += + 0.000000001
    for index_image in range(0,RealizationAmountPerSet):
        KLDivergence = 0.5 * (scipy.stats.entropy(ClusterDistribution_TI[scaleIndex],ClusterDistribution_Realization_PatternCTDS[scaleIndex][index_image]) + scipy.stats.entropy(ClusterDistribution_Realization_PatternCTDS[scaleIndex][index_image],ClusterDistribution_TI[scaleIndex]))
        KLDivergenceMatrix_PatternCTDS[scaleIndex][0][index_image+1] = KLDivergence
        KLDivergenceMatrix_PatternCTDS[scaleIndex][index_image+1][0] = KLDivergence
    for point_y in range(0,RealizationAmountPerSet):
        for point_x in range(point_y+1,RealizationAmountPerSet):
            KLDivergence = 0.5 * (scipy.stats.entropy(ClusterDistribution_Realization_PatternCTDS[scaleIndex][point_y],ClusterDistribution_Realization_PatternCTDS[scaleIndex][point_x]) + scipy.stats.entropy(ClusterDistribution_Realization_PatternCTDS[scaleIndex][point_x],ClusterDistribution_Realization_PatternCTDS[scaleIndex][point_y]))
            KLDivergenceMatrix_PatternCTDS[scaleIndex][point_y+1][point_x+1] = KLDivergence
            KLDivergenceMatrix_PatternCTDS[scaleIndex][point_x+1][point_y+1] = KLDivergence
    # print('the shape of distance matrix is '+str(KLDivergenceMatrix_PatternCTDS[scaleIndex].shape))
    # print('the distance with TI is '+str(KLDivergenceMatrix_PatternCTDS[scaleIndex][0]))
    print('the mean distance between Pattern CTDS realizations and TI is '+str(np.mean(KLDivergenceMatrix_PatternCTDS[scaleIndex][0])))
```

```python
# output the quantitative rank
distanceAmongRealizations_DS = 0
distanceAmongRealizations_PointCTS = 0
distanceAmongRealizations_PatternCTS = 0
distanceToTI_DS = 0
distanceToTI_PointDS = 0
distanceToTI_PatternDS = 0

for scaleIndex in range(0,ScaleIndex_Max+1):
    scaleZoom = 2**scaleIndex
    
    distanceAmongRealizations_DS += (1/scaleZoom)*(np.mean(KLDivergenceMatrix_DS[scaleIndex][1:RealizationAmountPerSet,1:RealizationAmountPerSet]))
    distanceAmongRealizations_PointCTS += (1/scaleZoom)*(np.mean(KLDivergenceMatrix_PointCTDS[scaleIndex][1:RealizationAmountPerSet,1:RealizationAmountPerSet]))
    distanceAmongRealizations_PatternCTS += (1/scaleZoom)*(np.mean(KLDivergenceMatrix_PatternCTDS[scaleIndex][1:RealizationAmountPerSet,1:RealizationAmountPerSet]))

    distanceToTI_DS += (1/scaleZoom)*(np.mean(KLDivergenceMatrix_DS[scaleIndex][0]))
    distanceToTI_PointDS += (1/scaleZoom)*(np.mean(KLDivergenceMatrix_PointCTDS[scaleIndex][0]))
    distanceToTI_PatternDS += (1/scaleZoom)*(np.mean(KLDivergenceMatrix_PatternCTDS[scaleIndex][0]))

distanceAmong_PointCTS_DS = distanceAmongRealizations_PointCTS/distanceAmongRealizations_DS
distanceAmong_PatternCTS_DS = distanceAmongRealizations_PatternCTS/distanceAmongRealizations_DS
print('the distance among realizations: Point CTS versus DS is '+str(distanceAmong_PointCTS_DS) +' : 1.00')
print('the distance among realizations: Pattern CTS versus DS is '+str(distanceAmong_PatternCTS_DS) +' : 1.00')

distanceBetween_PointCTS_DS = distanceToTI_PointDS/distanceToTI_DS
distanceBetween_PatternCTS_DS = distanceToTI_PatternDS/distanceToTI_DS
print('the distance to TI: PointCTS versus DS is '+str(distanceBetween_PointCTS_DS) +' : 1.00')
print('the distance to TI: PatternCTS versus DS is '+str(distanceBetween_PatternCTS_DS) +' : 1.00')

print('the comprehensive ratio: Point CTS versus DS is '+str(distanceAmong_PointCTS_DS/distanceBetween_PointCTS_DS)+' : 1.00')
print('the comprehensive ratio: Pattern CTS versus DS is '+str(distanceAmong_PatternCTS_DS/distanceBetween_PatternCTS_DS)+' : 1.00')
```

```python
# calculate the distance matrix at each scale   
# the core task is to compare DS and TDS realizations
                   
ClusterAmount = ClusterAmount_Max

KLDivergenceMatrix_Global_multiScare = np.zeros((ScaleIndex_Max+1,3*RealizationAmountPerSet+1,3*RealizationAmountPerSet+1))

for scaleIndex in range(0,ScaleIndex_Max+1):
    ClusterDistribution_Realization_Global = np.zeros((3*RealizationAmountPerSet+1,ClusterAmount))
    ClusterDistribution_Realization_Global[0] = ClusterDistribution_TI[scaleIndex]
    for realizationIndex in range(0,RealizationAmountPerSet):
        ClusterDistribution_Realization_Global[3*realizationIndex+1] = ClusterDistribution_Realization_DS[scaleIndex][realizationIndex]
        ClusterDistribution_Realization_Global[3*realizationIndex+2] = ClusterDistribution_Realization_PointCTDS[scaleIndex][realizationIndex]
        ClusterDistribution_Realization_Global[3*realizationIndex+3] = ClusterDistribution_Realization_PatternCTDS[scaleIndex][realizationIndex]

    for point_y in range(0,3*RealizationAmountPerSet+1):
        for point_x in range(point_y+1,3*RealizationAmountPerSet+1):
            KLDivergence = 0.5 * (scipy.stats.entropy(ClusterDistribution_Realization_Global[point_y],ClusterDistribution_Realization_Global[point_x]) + scipy.stats.entropy(ClusterDistribution_Realization_Global[point_x],ClusterDistribution_Realization_Global[point_y]))
            KLDivergenceMatrix_Global_multiScare[scaleIndex][point_y][point_x] = KLDivergence
            KLDivergenceMatrix_Global_multiScare[scaleIndex][point_x][point_y] = KLDivergence
```

```python
# calculate the overall distance
MDSVariable_Maximum = 2
mds = manifold.MDS(n_components=MDSVariable_Maximum, max_iter=3000, eps=1e-9, dissimilarity="precomputed", n_jobs=1)

KLDivergenceMatrix_Global = np.zeros((3*RealizationAmountPerSet+1,3*RealizationAmountPerSet+1))
for point_y in range(0,3*RealizationAmountPerSet+1):
    for point_x in range(0,3*RealizationAmountPerSet+1):
        distance = 0.0
        for scaleIndex in range(0,ScaleIndex_Max+1):
            scaleZoom = 2**scaleIndex
            distance += (1/scaleZoom)*(KLDivergenceMatrix_Global_multiScare[scaleIndex][point_y][point_x])
        KLDivergenceMatrix_Global[point_y][point_x] = distance
```

```python
# conduct the MDS
pos_Global = mds.fit(KLDivergenceMatrix_Global).embedding_
```

```python
# output the MDS diagram
for realizationIndex in range(0,RealizationAmountPerSet):
    plt.scatter(pos_Global[3*realizationIndex+1, 0], pos_Global[3*realizationIndex+1, 1], color='forestgreen', s=30, lw=0)
    plt.scatter(pos_Global[3*realizationIndex+2, 0], pos_Global[3*realizationIndex+2, 1], color='darkmagenta', s=30, lw=0)
    plt.scatter(pos_Global[3*realizationIndex+3, 0], pos_Global[3*realizationIndex+3, 1], color='deepskyblue', s=30, lw=0)
plt.scatter(pos_Global[1, 0], pos_Global[1, 1], color='forestgreen', s=20, lw=0, label='DS')
plt.scatter(pos_Global[2, 0], pos_Global[2, 1], color='darkmagenta', s=20, lw=0, label='TDS-pixel')
plt.scatter(pos_Global[3, 0], pos_Global[3, 1], color='deepskyblue', s=20, lw=0, label='TDS-patch')
plt.scatter(pos_Global[0, 0], pos_Global[0, 1], color='red', s=70, lw=0, label='TI')
print('the size of pos_DS is '+str(pos_Global.shape))
plt.legend(scatterpoints=1, loc='upper right', shadow=False,fontsize='large')
plt.xlabel(str('MDS1'),fontsize='large')
plt.ylabel(str('MDS2'),fontsize='large')
plt.tick_params(direction='in',labelsize='large')
plt.xlim(-0.11,0.22)
plt.ylim(-0.11,0.11)
plt.xticks(np.linspace(start=-0.1,stop=0.20,num=7))
plt.yticks(np.linspace(start=-0.1,stop=0.10,num=5))
plt.show()
```
