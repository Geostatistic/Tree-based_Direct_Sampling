```python
import numpy as np
import matplotlib.pyplot as plt

from scipy import ndimage
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
# the shape of realizations
Realization_Height = 1202
Realization_Width = 1202

RoutingModelAmount = 3  # the amount of routing model for one algorithm


# this range is used to determine the range of bins in histogram
binsRange = np.arange(start=30000,stop=170001,step=10000)
```

```python
# read the simulation area and find the boundary
from scipy import interpolate

SimulationArea_row = np.loadtxt('AntarcticaRealization\Antarctica_ds_data4.txt',skiprows=3)
# print(SimulationArea_row)
# print('the size of realization is '+str(SimulationArea_row.shape))
SimulationArea_2D = SimulationArea_row.reshape(Realization_Height//2,Realization_Width//2)
# plt.imshow(SimulationArea_2D,cmap = 'jet', vmin=-7000, vmax=3346)
# plt.title('ds_data4')
# plt.colorbar()
# plt.show()

# 2D interpolation
ZoomCoefficient = 2
coordinate_y = np.linspace(0,Realization_Height-ZoomCoefficient,int(Realization_Height/ZoomCoefficient))
coordinate_x = np.linspace(0,Realization_Width-ZoomCoefficient,int(Realization_Width/ZoomCoefficient))
coordinate_y_new = np.linspace(0,Realization_Height-1,Realization_Height)
coordinate_x_new = np.linspace(0,Realization_Width-1,Realization_Width)

interpolationFunction = interpolate.interp2d(coordinate_x, coordinate_y, SimulationArea_2D, kind='linear')
SimulationArea_2D = interpolationFunction(coordinate_x_new, coordinate_y_new)
SimulationArea_row = SimulationArea_2D.reshape(Realization_Height*Realization_Width)
# print('the size of Realization_2D matrix is '+str(SimulationArea_2D.shape))
# print('the size of realization_row is '+str(SimulationArea_row.shape))

# plt.imshow(SimulationArea_2D,cmap = 'jet', vmin=-7000, vmax=3346)
# plt.title('ds_data4 After interpolation')
# plt.colorbar()
# plt.show()

Boundary_2D = np.zeros((Realization_Height, Realization_Width))
Boundary_2D[ SimulationArea_2D < -7000 ] = 1.0
SimulationArea_2D = np.zeros((Realization_Height, Realization_Width))
for y in range(1,Realization_Height-1):
    for x in range(1,Realization_Width-1):
        if(Boundary_2D[y][x] == 1):
            for i in range(-1,2):
                for j in range(-1,2):
                    if(Boundary_2D[y+i][x+j] == 0):
                        SimulationArea_2D[y][x] = 1
                     
Boundary_2D = Boundary_2D.T
SimulationArea_2D = SimulationArea_2D.T

# plt.imshow(Boundary_2D,cmap = 'jet')
# plt.title('SimulationArea area')
# plt.colorbar()
# plt.show()

Boundary_2D = np.copy(SimulationArea_2D)
Boundary_2D = Boundary_2D*-1 + 1
# plt.imshow(Boundary_2D,cmap = 'jet')
# plt.title('Boundary area')
# plt.colorbar()
# plt.show()

Boundary_y = []
Boundary_x = []
for y in range(1,Realization_Height-1):
    for x in range(1,Realization_Width-1):
        if(Boundary_2D[y][x] == 0):
            Boundary_y.append(y)
            Boundary_x.append(x)
```

```python
RoutingModel_TDS = np.zeros((RoutingModelAmount,Realization_Height,Realization_Width))

for routingModelIndex in range(0,RoutingModelAmount):
    print("TDS routing model "+str(routingModelIndex));
    Output = np.loadtxt('AntarcticaRealization\z_RoutingModel_z_Antarctica_PatternTDS'+str(routingModelIndex)+'.txt')
    Output = Output.reshape(Realization_Height,Realization_Width)
    RoutingModel_TDS[routingModelIndex] = np.transpose(Output)
```

```python
# perform max filter
for routingModelIndex in range(0,RoutingModelAmount):
    output = ndimage.maximum_filter(RoutingModel_TDS[routingModelIndex],size=10)
    print('TDS water routing model '+str(routingModelIndex))
    plt.figure(figsize=(7,7))
    plt.imshow(output,cmap = 'hot',vmin=0,vmax=140000)
    # plt.title('TDS routing model'+str(routingModelIndex))
    plt.title('flow accumulation based on the TDS realization',fontsize='x-large')
    plt.tick_params(direction='in',labelsize='large')
    plt.colorbar(fraction=0.046,pad=0.06,ticks=np.arange(0,140001,step=20000))
    plt.show()
```

```python
# perform max filter with boundary
for routingModelIndex in range(0,RoutingModelAmount):
    output = ndimage.maximum_filter(RoutingModel_TDS[routingModelIndex],size=10)
    print('TDS water routing model '+str(routingModelIndex))
    plt.figure(figsize=(7,7))
    plt.imshow(output,cmap = 'hot',vmin=0,vmax=140000)
    # plt.title('TDS routing model'+str(routingModelIndex))
    plt.title('flow accumulation based on the TDS realization',fontsize='x-large')
    plt.tick_params(direction='in',labelsize='large')
    plt.colorbar(fraction=0.046,pad=0.06,ticks=np.arange(0,140001,step=20000))
    
    plt.plot(Boundary_x, Boundary_y, 'c.',linewidth=0.1,markersize=0.3)
    
    plt.show()
```

```python
# output the histogram
for routingModelIndex in range(0,RoutingModelAmount):    
    output = RoutingModel_TDS[routingModelIndex].reshape(Realization_Height*Realization_Width)
    plt.hist(output,bins=binsRange,rwidth=0.6)
    #plt.xticks(binsRange+10000)
    plt.tick_params(direction='in')
    plt.title('histogram of TDS routing model '+str(routingModelIndex))
    plt.show()
```

```python
RoutingModel_DS = np.zeros((RoutingModelAmount,Realization_Height,Realization_Width))

for routingModelIndex in range(0,RoutingModelAmount):
    print("TDS routing model "+str(routingModelIndex));
    Output = np.loadtxt('AntarcticaRealization\z_RoutingModel_z_Antarctica_DS'+str(routingModelIndex)+'.txt')
    Output = Output.reshape(Realization_Height,Realization_Width)
    RoutingModel_DS[routingModelIndex] = np.transpose(Output)
```

```python
# perform max filter
for routingModelIndex in range(0,RoutingModelAmount):
    output = ndimage.maximum_filter(RoutingModel_DS[routingModelIndex],size=10)
    print('DS water routing model '+str(routingModelIndex))
    plt.figure(figsize=(7,7))
    plt.imshow(output,cmap = 'hot',vmin=0,vmax=140000)
    # plt.title('TDS routing model'+str(routingModelIndex))
    plt.title('flow accumulation based on the DS realization',fontsize='x-large')
    plt.tick_params(direction='in',labelsize='large')
    plt.colorbar(fraction=0.046,pad=0.06,ticks=np.arange(0,140001,step=20000))
    plt.show()
```

```python
# perform max filter with boundary
for routingModelIndex in range(0,RoutingModelAmount):
    output = ndimage.maximum_filter(RoutingModel_DS[routingModelIndex],size=10)
    print('DS water routing model '+str(routingModelIndex))
    plt.figure(figsize=(7,7))
    plt.imshow(output,cmap = 'hot',vmin=0,vmax=140000)
    # plt.title('TDS routing model'+str(routingModelIndex))
    plt.title('flow accumulation based on the DS realization',fontsize='x-large')
    plt.tick_params(direction='in',labelsize='large')
    plt.colorbar(fraction=0.046,pad=0.06,ticks=np.arange(0,140001,step=20000))
    plt.plot(Boundary_x, Boundary_y, 'c.',linewidth=0.1,markersize=0.3)
    plt.show()
```

```python
for routingModelIndex in range(0,RoutingModelAmount):    
    output = RoutingModel_DS[routingModelIndex].reshape(Realization_Height*Realization_Width)
    plt.hist(output,bins=binsRange,rwidth=0.6)
    #plt.xticks(binsRange+10000)
    plt.tick_params(direction='in')
    plt.title('histogram of DS routing model '+str(routingModelIndex))
    plt.show()
```

```python
# output the histogram   the y value is the occurrence
output = RoutingModel_TDS.reshape(RoutingModelAmount*Realization_Height*Realization_Width)
plt.hist(output,bins=binsRange,rwidth=0.6)
plt.tick_params(direction='in')
plt.yticks(np.linspace(0,16000,9))
plt.title('histogram of TDS routing models')
plt.show()

output = RoutingModel_DS.reshape(RoutingModelAmount*Realization_Height*Realization_Width)
plt.hist(output,bins=binsRange,rwidth=0.6)
plt.tick_params(direction='in')
plt.yticks(np.linspace(0,16000,9))
plt.title('histogram of DS routing models')
plt.show()
```

```python
# output the histogram the y value is the frequency
hist, bin_edges = np.histogram(RoutingModel_TDS,bins=binsRange)
hist = hist/(Realization_Height*Realization_Width*RoutingModelAmount)
print('TDS Frequency:')
for number in hist:
    print(number)
    
# output the histogram the y value is the frequency
hist, bin_edges = np.histogram(RoutingModel_DS,bins=binsRange)
hist = hist/(Realization_Height*Realization_Width*RoutingModelAmount)
print('DS Frequency:')
for number in hist:
    print(number)
```

```python
meanRouting_TDS = np.zeros((Realization_Height,Realization_Width))
varianceRoutingTDS = np.zeros((Realization_Height,Realization_Width))

meanRouting_TDS = np.mean(RoutingModel_TDS,axis=0)
print('the size of mean is '+str(meanRouting_TDS.shape))

varianceRouting_TDS = np.var(RoutingModel_TDS,axis=0)
print('the size of variance is '+str(varianceRouting_TDS.shape))

meanRouting_DS = np.mean(RoutingModel_DS,axis=0)
varianceRouting_DS = np.var(RoutingModel_DS,axis=0)
```

```python
# perform max filter on mean water routing models
output = ndimage.maximum_filter(meanRouting_TDS,size=10)
print('mean of TDS water routing models')
plt.figure(figsize=(7,7))
plt.imshow(output,cmap = 'hot',vmin=0,vmax=140000)
# plt.title('TDS routing model'+str(routingModelIndex))
plt.title('mean of flow accumulation based on TDS realizations',fontsize='x-large')
plt.tick_params(direction='in',labelsize='large')
plt.colorbar(fraction=0.046,pad=0.06,ticks=np.arange(0,140001,step=20000))
plt.plot(Boundary_x, Boundary_y, 'c.',linewidth=0.1,markersize=0.3)
plt.show()

output = ndimage.maximum_filter(meanRouting_DS,size=10)
print('mean of DS water routing models')
plt.figure(figsize=(7,7))
plt.imshow(output,cmap = 'hot',vmin=0,vmax=140000)
# plt.title('TDS routing model'+str(routingModelIndex))
plt.title('mean of flow accumulation based on DS realizations',fontsize='x-large')
plt.tick_params(direction='in',labelsize='large')
plt.colorbar(fraction=0.046,pad=0.06,ticks=np.arange(0,140001,step=20000))
plt.plot(Boundary_x, Boundary_y, 'c.',linewidth=0.1,markersize=0.3)
plt.show()
```

```python
# perform max filter on variance water routing models
output = ndimage.maximum_filter(varianceRouting_TDS,size=10)
print('variance of TDS water routing models')
plt.figure(figsize=(7,7))
plt.imshow(output,cmap = 'hot',vmin=0,vmax=2e9)
# plt.title('TDS routing model'+str(routingModelIndex))
plt.title('variance of flow accumulation based on TDS realizations',fontsize='x-large')
plt.tick_params(direction='in',labelsize='large')
# plt.colorbar()
plt.colorbar(fraction=0.046,pad=0.1,ticks=np.arange(0,2e9+1,step=2e8))
plt.show()

output = ndimage.maximum_filter(varianceRouting_DS,size=10)
print('variance of DS water routing models')
plt.figure(figsize=(7,7))
plt.imshow(output,cmap = 'hot',vmin=0,vmax=2e9)
# plt.title('TDS routing model'+str(routingModelIndex))
plt.title('variance of flow accumulation based on DS realizations',fontsize='x-large')
plt.tick_params(direction='in',labelsize='large')
# plt.colorbar()
plt.colorbar(fraction=0.046,pad=0.1,ticks=np.arange(0,2e9+1,step=2e8))
plt.show()
```
