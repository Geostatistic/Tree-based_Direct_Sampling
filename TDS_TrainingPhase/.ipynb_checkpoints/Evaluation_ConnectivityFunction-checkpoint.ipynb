```python
# set the number of realizations
RealizationAmountPerSet = 20

# set the parameter of variogram
Connectivity_Direction_X = 1
Connectivity_Direction_Y = 0
Connectivity_MaxDistance = 35

TargetFacies = 1
```

```python
import matplotlib.pyplot as plt
import numpy as np

from PIL import Image

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
# set some temporary variable
point_y = 0
point_x = 0
point_value = 0

sample_y = 0
sample_x = 0
sample_value = 0
```

```python
TIBMP = Image.open("ChannelRealizations\channel101.bmp",'r')

TIInt = TIBMP.load()
TI_Width = TIBMP.size[0]
TI_Height = TIBMP.size[1]
TI = np.empty((TI_Height,TI_Width))

for point_y in range(TI_Height):
    for point_x in range(TI_Width):
        point_value = float(TIInt[point_x,point_y][0])
        TI[point_y][point_x] = point_value/255;

print(TI_Width)
print(TI_Height)
# the horizontal direction is x-axis and corresponds to the width
# the vertical direction is y-axis and correspongds to the height
print("the training image is: ")
plt.imshow(TI,vmin=0, vmax=1)
plt.colorbar()  # show color scale
plt.show()
```

```python
# read the DS realization
Realization_DS = np.empty((RealizationAmountPerSet,TI_Height,TI_Width))
index_image = 0;
for index_image in range(0,RealizationAmountPerSet):
    TIBMP = Image.open("ChannelRealizations\Realization_DS"+str(index_image)+".bmp",'r')
    TIInt = TIBMP.load()
    for point_y in range(TI_Height):
        for point_x in range(TI_Width):
            point_value = float(TIInt[point_x,point_y][0])
            Realization_DS[index_image][point_y][point_x] = point_value/255;
print("the first realization is: ")
plt.imshow(Realization_DS[0],vmin=0, vmax=1)
plt.colorbar()  # show color scale
plt.show()
```

```python
# read the CTS realization
Realization_PointCTS = np.empty((RealizationAmountPerSet,TI_Height,TI_Width))
for index_image in range(0,RealizationAmountPerSet):
    TIBMP = Image.open("ChannelRealizations\Realization_PointCTS"+str(index_image)+".bmp",'r')
    TIInt = TIBMP.load()
    for point_y in range(TI_Height):
        for point_x in range(TI_Width):
            point_value = float(TIInt[point_x,point_y][0])
            Realization_PointCTS[index_image][point_y][point_x] = point_value/255;
print("the first realization is: ")
plt.imshow(Realization_PointCTS[0],vmin=0, vmax=1)
plt.colorbar()  # show color scale
plt.show()
```

```python
# read the pattern-based cluster tree simulation realization
Realization_PatternCTS = np.empty((RealizationAmountPerSet,TI_Height,TI_Width))
for index_image in range(0,RealizationAmountPerSet):
    TIBMP = Image.open("ChannelRealizations\Realization_PatternCTS"+str(index_image)+".bmp",'r')
    TIInt = TIBMP.load()
    for point_y in range(TI_Height):
        for point_x in range(TI_Width):
            point_value = float(TIInt[point_x,point_y][0])
            Realization_PatternCTS[index_image][point_y][point_x] = point_value/255;
print("the first realization is: ")
plt.imshow(Realization_PatternCTS[0],vmin=0, vmax=1)
plt.colorbar()  # show color scale
plt.show()
```

```python
# calculate the TI variogram
sampleAmount = int(TI_Height * TI_Width * 0.5)

Connectivity_Index = np.linspace(start=0,stop=Connectivity_MaxDistance-1,num=Connectivity_MaxDistance).astype(int)
Connectivity_TI = np.zeros((Connectivity_MaxDistance))

endPoint_relative_X = 0
endPoint_relative_Y = 0
pairDistance= 0.0
isConnective = False

while(True):
    # print('current relative coordinates: X: '+str(endPoint_relative_X)+' Y: '+str(endPoint_relative_Y))
    
    randomSample_X = np.random.randint(low=0,high=int(TI_Width-pairDistance-1),size=sampleAmount)
    randomSample_Y = np.random.randint(low=0,high=int(TI_Height-pairDistance-1),size=sampleAmount)
    ConnectivityElement = 0
    
    for point_x, point_y in zip(randomSample_X,randomSample_Y):
        isConnective = True
        sample_x = point_x
        sample_y = point_y
        for checkRadius in range(0,(int)(pairDistance)+1):
            sample_value = TI[sample_y][sample_x]
            if(sample_value == TargetFacies):
                sample_x += Connectivity_Direction_X
                sample_y += Connectivity_Direction_Y
                continue
            else:
                isConnective = False
                break
        if(isConnective == True):
            ConnectivityElement += 1
    ConnectivityElement = ConnectivityElement/sampleAmount
    Connectivity_TI[int(pairDistance+0.5)] = ConnectivityElement

    endPoint_relative_X += Connectivity_Direction_X
    endPoint_relative_Y += Connectivity_Direction_Y
    pairDistance = np.sqrt(endPoint_relative_X**2 + endPoint_relative_Y**2)
    # print('pair distance is '+str(pairDistance))
    if(pairDistance >= Connectivity_MaxDistance):
        break
    
fix, ax = plt.subplots()
ax.plot(Connectivity_Index,Connectivity_TI,color='red',linewidth=2,label='connectivity of TI')
ax.legend()
plt.show()

# print('the variogram of TI is ')
# for variogramElement in Variogram_TI:
#     print(variogramElement)
```

```python
# calculate the DS realization variogram
Connectivity_DS = np.zeros((RealizationAmountPerSet,Connectivity_MaxDistance))

for index_image in range(0,RealizationAmountPerSet):
    endPoint_relative_X = 0
    endPoint_relative_Y = 0
    pairDistance= 0.0

    while(True):
        randomSample_X = np.random.randint(low=0,high=int(TI_Width-pairDistance-1),size=sampleAmount)
        randomSample_Y = np.random.randint(low=0,high=int(TI_Height-pairDistance-1),size=sampleAmount)
        ConnectivityElement = 0

        for point_x, point_y in zip(randomSample_X,randomSample_Y):
            isConnective = True
            sample_x = point_x
            sample_y = point_y
            for checkRadius in range(0,(int)(pairDistance)+1):
                sample_value = Realization_DS[index_image][sample_y][sample_x]
                if(sample_value == TargetFacies):
                    sample_x += Connectivity_Direction_X
                    sample_y += Connectivity_Direction_Y
                    continue
                else:
                    isConnective = False
                    break
            if(isConnective == True):
                ConnectivityElement += 1
        ConnectivityElement = ConnectivityElement/sampleAmount
        Connectivity_DS[index_image][int(pairDistance+0.5)] = ConnectivityElement

        endPoint_relative_X += Connectivity_Direction_X
        endPoint_relative_Y += Connectivity_Direction_Y
        pairDistance = np.sqrt(endPoint_relative_X**2 + endPoint_relative_Y**2)
        if(pairDistance >= Connectivity_MaxDistance):
            break
```

```python
# calculate the point TDS realization variogram
Connectivity_PointCTS = np.zeros((RealizationAmountPerSet,Connectivity_MaxDistance))

for index_image in range(0,RealizationAmountPerSet):
    endPoint_relative_X = 0
    endPoint_relative_Y = 0
    pairDistance= 0.0

    while(True):
        randomSample_X = np.random.randint(low=0,high=int(TI_Width-pairDistance-1),size=sampleAmount)
        randomSample_Y = np.random.randint(low=0,high=int(TI_Height-pairDistance-1),size=sampleAmount)
        ConnectivityElement = 0

        for point_x, point_y in zip(randomSample_X,randomSample_Y):
            isConnective = True
            sample_x = point_x
            sample_y = point_y
            for checkRadius in range(0,(int)(pairDistance)+1):
                sample_value = Realization_PointCTS[index_image][sample_y][sample_x]
                if(sample_value == TargetFacies):
                    sample_x += Connectivity_Direction_X
                    sample_y += Connectivity_Direction_Y
                    continue
                else:
                    isConnective = False
                    break
            if(isConnective == True):
                ConnectivityElement += 1
        ConnectivityElement = ConnectivityElement/sampleAmount
        Connectivity_PointCTS[index_image][int(pairDistance+0.5)] = ConnectivityElement

        endPoint_relative_X += Connectivity_Direction_X
        endPoint_relative_Y += Connectivity_Direction_Y
        pairDistance = np.sqrt(endPoint_relative_X**2 + endPoint_relative_Y**2)
        if(pairDistance >= Connectivity_MaxDistance):
            break
```

```python
# calculate the pattern TDS realization variogram
Connectivity_PatternCTS = np.zeros((RealizationAmountPerSet,Connectivity_MaxDistance))

for index_image in range(0,RealizationAmountPerSet):
    endPoint_relative_X = 0
    endPoint_relative_Y = 0
    pairDistance= 0.0

    while(True):
        randomSample_X = np.random.randint(low=0,high=int(TI_Width-pairDistance-1),size=sampleAmount)
        randomSample_Y = np.random.randint(low=0,high=int(TI_Height-pairDistance-1),size=sampleAmount)
        ConnectivityElement = 0

        for point_x, point_y in zip(randomSample_X,randomSample_Y):
            isConnective = True
            sample_x = point_x
            sample_y = point_y
            for checkRadius in range(0,(int)(pairDistance)+1):
                sample_value = Realization_PatternCTS[index_image][sample_y][sample_x]
                if(sample_value == TargetFacies):
                    sample_x += Connectivity_Direction_X
                    sample_y += Connectivity_Direction_Y
                    continue
                else:
                    isConnective = False
                    break
            if(isConnective == True):
                ConnectivityElement += 1
        ConnectivityElement = ConnectivityElement/sampleAmount
        Connectivity_PatternCTS[index_image][int(pairDistance+0.5)] = ConnectivityElement

        endPoint_relative_X += Connectivity_Direction_X
        endPoint_relative_Y += Connectivity_Direction_Y
        pairDistance = np.sqrt(endPoint_relative_X**2 + endPoint_relative_Y**2)
        if(pairDistance >= Connectivity_MaxDistance):
            break
```

```python
# draw the variogram diagram respectively
fix, ax = plt.subplots()
ax.plot(Connectivity_Index,Connectivity_DS[0],color='green',linewidth=0.5,label='DS')
for index_image in range(1,RealizationAmountPerSet):
    ax.plot(Connectivity_Index,Connectivity_DS[index_image],color='green',linewidth=0.5)
ax.plot(Connectivity_Index,Connectivity_TI,color='red',linewidth=2,label='TI')
ax.legend(loc='upper right',fontsize='x-large')
ax.set_xlabel('Lag distance',fontsize='x-large')
ax.set_ylabel('Connectivity Probability',fontsize='x-large')
ax.tick_params(direction='in',labelsize=12)
plt.show()

fix, ax = plt.subplots()
ax.plot(Connectivity_Index,Connectivity_PointCTS[0],color='darkmagenta',linewidth=0.5,label='TDS-pixel')
for index_image in range(1,RealizationAmountPerSet):
    ax.plot(Connectivity_Index,Connectivity_PointCTS[index_image],color='darkmagenta',linewidth=0.5)
ax.plot(Connectivity_Index,Connectivity_TI,color='red',linewidth=2,label='TI')
ax.legend(loc='upper right',fontsize='x-large')
ax.set_xlabel('Lag distance',fontsize='x-large')
ax.set_ylabel('Connectivity Probability',fontsize='x-large')
ax.tick_params(direction='in',labelsize=12)
plt.show()

fix, ax = plt.subplots()
ax.plot(Connectivity_Index,Connectivity_PatternCTS[0],color='deepskyblue',linewidth=0.5,label='TDS-patch')
for index_image in range(1,RealizationAmountPerSet):
    ax.plot(Connectivity_Index,Connectivity_PatternCTS[index_image],color='deepskyblue',linewidth=0.5)
ax.plot(Connectivity_Index,Connectivity_TI,color='red',linewidth=2,label='TI')
ax.legend(loc='upper right',fontsize='x-large')
ax.set_xlabel('Lag distance',fontsize='x-large')
ax.set_ylabel('Connectivity Probability',fontsize='x-large')
ax.tick_params(direction='in',labelsize=12)
plt.show()
```
