```python
# set the number of realizations

# set the parameter of variogram
Variogram_Direction_X = 1
Variogram_Direction_Y = 0
Variogram_MaxDistance = 35
```

```python
import matplotlib.pyplot as plt
import numpy as np

from PIL import Image

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
# set some temporary variable
point_y = 0
point_x = 0
point_value = 0

sample_y = 0
sample_x = 0
sample_value = 0
```

```python
TIBMP = Image.open("ChannelRealizations\channel101.bmp",'r')

TIInt = TIBMP.load()
TI_Width = TIBMP.size[0]
TI_Height = TIBMP.size[1]
TI = np.empty((TI_Height,TI_Width))

for point_y in range(TI_Height):
    for point_x in range(TI_Width):
        point_value = float(TIInt[point_x,point_y][0])
        TI[point_y][point_x] = point_value/255;

print(TI_Width)
print(TI_Height)
# the horizontal direction is x-axis and corresponds to the width
# the vertical direction is y-axis and correspongds to the height
print("the training image is: ")
plt.imshow(TI,vmin=0, vmax=1)
plt.colorbar()  # show color scale
plt.show()
```

```python
# read the DS realization
Realization_DS = np.empty((RealizationAmountPerSet,TI_Height,TI_Width))
index_image = 0;
for index_image in range(0,RealizationAmountPerSet):
    TIBMP = Image.open("ChannelRealizations\Realization_DS"+str(index_image)+".bmp",'r')
    TIInt = TIBMP.load()
    for point_y in range(TI_Height):
        for point_x in range(TI_Width):
            point_value = float(TIInt[point_x,point_y][0])
            Realization_DS[index_image][point_y][point_x] = point_value/255;
print("the first realization is: ")
plt.imshow(Realization_DS[0],vmin=0, vmax=1)
plt.colorbar()  # show color scale
plt.show()
```

```python
# read the CTS realization
Realization_PointCTS = np.empty((RealizationAmountPerSet,TI_Height,TI_Width))
for index_image in range(0,RealizationAmountPerSet):
    TIBMP = Image.open("ChannelRealizations\Realization_PointCTS"+str(index_image)+".bmp",'r')
    TIInt = TIBMP.load()
    for point_y in range(TI_Height):
        for point_x in range(TI_Width):
            point_value = float(TIInt[point_x,point_y][0])
            Realization_PointCTS[index_image][point_y][point_x] = point_value/255;
print("the first realization is: ")
plt.imshow(Realization_PointCTS[0],vmin=0, vmax=1)
plt.colorbar()  # show color scale
plt.show()
```

```python
# read the pattern-based cluster tree simulation realization
Realization_PatternCTS = np.empty((RealizationAmountPerSet,TI_Height,TI_Width))
for index_image in range(0,RealizationAmountPerSet):
    TIBMP = Image.open("ChannelRealizations\Realization_PatternCTS"+str(index_image)+".bmp",'r')
    TIInt = TIBMP.load()
    for point_y in range(TI_Height):
        for point_x in range(TI_Width):
            point_value = float(TIInt[point_x,point_y][0])
            Realization_PatternCTS[index_image][point_y][point_x] = point_value/255;
print("the first realization is: ")
plt.imshow(Realization_PatternCTS[0],vmin=0, vmax=1)
plt.colorbar()  # show color scale
plt.show()
```

```python
# calculate the TI variogram
pairDistance = 0.0
relativeCoordinate_X = 0
relativeCoordinate_Y = 0

sampleAmount = int(TI_Height * TI_Width * 0.5)

print('the variance of Training Image is '+str(np.var(TI)))

Variogram_Index = np.linspace(start=0,stop=Variogram_MaxDistance-1,num=Variogram_MaxDistance).astype(int)
print('the x axis in variogram is '+str(Variogram_Index))
Variogram_TI = np.zeros((Variogram_MaxDistance))
print('the initial variogram is '+str(Variogram_TI))

# # just used for test
# randomSample_X = np.random.randint(low=0,high=int(TI_Width-pairDistance-1),size=sampleAmount)
# print(randomSample_X)
# print('the shape of randomSample is '+str(randomSample_X.shape))
# print('the maximum in random set is '+str(np.max(randomSample_X)))

while(True):
    relativeCoordinate_X += Variogram_Direction_X
    relativeCoordinate_Y += Variogram_Direction_Y
    pairDistance = np.sqrt(relativeCoordinate_X**2 + relativeCoordinate_Y**2)
    if(pairDistance >= Variogram_MaxDistance):
        break
    # print('current relative coordinates: X: '+str(relativeCoordinate_X)+' Y: '+str(relativeCoordinate_Y))
    
    variogramElement = 0.0
    randomSample_X = np.random.randint(low=0,high=int(TI_Width-pairDistance-1),size=sampleAmount)
    randomSample_Y = np.random.randint(low=0,high=int(TI_Height-pairDistance-1),size=sampleAmount)
    
    for point_x, point_y in zip(randomSample_X,randomSample_Y):
        sample_x = point_x + relativeCoordinate_X
        sample_y = point_y + relativeCoordinate_Y
        
        point_value = TI[point_y][point_x]
        sample_value = TI[sample_y][sample_x]
        
        variogramElement += (point_value - sample_value)**2
    variogramElement = variogramElement/sampleAmount
    Variogram_TI[int(pairDistance+0.5)] = variogramElement

    
fix, ax = plt.subplots()
ax.plot(Variogram_Index,Variogram_TI,color='red',linewidth=2,label='variogram of TI')
ax.legend()
plt.show()

# print('the variogram of TI is ')
# for variogramElement in Variogram_TI:
#     print(variogramElement)
```

```python
# calculate the DS realization variogram
Variogram_DS = np.zeros((RealizationAmountPerSet,Variogram_MaxDistance))

for index_image in range(0,RealizationAmountPerSet):
    relativeCoordinate_X = 0
    relativeCoordinate_Y = 0
    while(True):
        relativeCoordinate_X += Variogram_Direction_X
        relativeCoordinate_Y += Variogram_Direction_Y
        pairDistance = np.sqrt(relativeCoordinate_X**2 + relativeCoordinate_Y**2)
        if(pairDistance >= Variogram_MaxDistance):
            break

        variogramElement = 0.0
        randomSample_X = np.random.randint(low=0,high=int(TI_Width-pairDistance-1),size=sampleAmount)
        randomSample_Y = np.random.randint(low=0,high=int(TI_Height-pairDistance-1),size=sampleAmount)

        for point_x, point_y in zip(randomSample_X,randomSample_Y):
            sample_x = point_x + relativeCoordinate_X
            sample_y = point_y + relativeCoordinate_Y

            point_value = Realization_DS[index_image][point_y][point_x]
            sample_value = Realization_DS[index_image][sample_y][sample_x]

            variogramElement += (point_value - sample_value)**2
        variogramElement = variogramElement/sampleAmount
        Variogram_DS[index_image][int(pairDistance+0.5)] = variogramElement
```

```python
# calculate the point-based CTS realization variogram
Variogram_PointCTS = np.zeros((RealizationAmountPerSet,Variogram_MaxDistance))

for index_image in range(0,RealizationAmountPerSet):
    relativeCoordinate_X = 0
    relativeCoordinate_Y = 0
    while(True):
        relativeCoordinate_X += Variogram_Direction_X
        relativeCoordinate_Y += Variogram_Direction_Y
        pairDistance = np.sqrt(relativeCoordinate_X**2 + relativeCoordinate_Y**2)
        if(pairDistance >= Variogram_MaxDistance):
            break

        variogramElement = 0.0
        randomSample_X = np.random.randint(low=0,high=int(TI_Width-pairDistance-1),size=sampleAmount)
        randomSample_Y = np.random.randint(low=0,high=int(TI_Height-pairDistance-1),size=sampleAmount)

        for point_x, point_y in zip(randomSample_X,randomSample_Y):
            sample_x = point_x + relativeCoordinate_X
            sample_y = point_y + relativeCoordinate_Y

            point_value = Realization_PointCTS[index_image][point_y][point_x]
            sample_value = Realization_PointCTS[index_image][sample_y][sample_x]

            variogramElement += (point_value - sample_value)**2
        variogramElement = variogramElement/sampleAmount
        Variogram_PointCTS[index_image][int(pairDistance+0.5)] = variogramElement
     
```

```python
# calculate the pattern-based CTS realization variogram
Variogram_PatternCTS = np.zeros((RealizationAmountPerSet,Variogram_MaxDistance))

for index_image in range(0,RealizationAmountPerSet):
    relativeCoordinate_X = 0
    relativeCoordinate_Y = 0
    while(True):
        relativeCoordinate_X += Variogram_Direction_X
        relativeCoordinate_Y += Variogram_Direction_Y
        pairDistance = np.sqrt(relativeCoordinate_X**2 + relativeCoordinate_Y**2)
        if(pairDistance >= Variogram_MaxDistance):
            break

        variogramElement = 0.0
        randomSample_X = np.random.randint(low=0,high=int(TI_Width-pairDistance-1),size=sampleAmount)
        randomSample_Y = np.random.randint(low=0,high=int(TI_Height-pairDistance-1),size=sampleAmount)

        for point_x, point_y in zip(randomSample_X,randomSample_Y):
            sample_x = point_x + relativeCoordinate_X
            sample_y = point_y + relativeCoordinate_Y

            point_value = Realization_PatternCTS[index_image][point_y][point_x]
            sample_value = Realization_PatternCTS[index_image][sample_y][sample_x]

            variogramElement += (point_value - sample_value)**2
        variogramElement = variogramElement/sampleAmount
        Variogram_PatternCTS[index_image][int(pairDistance+0.5)] = variogramElement
```

```python
# draw the variogram diagram respectively
fix, ax = plt.subplots()
ax.plot(Variogram_Index,Variogram_DS[0],color='green',linewidth=0.5,label='variogram of DS')
for index_image in range(1,RealizationAmountPerSet):
    ax.plot(Variogram_Index,Variogram_DS[index_image],color='green',linewidth=0.5)
ax.plot(Variogram_Index,Variogram_TI,color='red',linewidth=2,label='variogram of TI')
ax.legend(loc='lower right',fontsize='xx-large')
plt.show()

fix, ax = plt.subplots()
ax.plot(Variogram_Index,Variogram_PointCTS[0],color='darkmagenta',linewidth=0.5,label='variogram of point TDS')
for index_image in range(1,RealizationAmountPerSet):
    ax.plot(Variogram_Index,Variogram_PointCTS[index_image],color='darkmagenta',linewidth=0.5)
ax.plot(Variogram_Index,Variogram_TI,color='red',linewidth=2,label='variogram of TI')
ax.legend(loc='lower right',fontsize='xx-large')
plt.show()

fix, ax = plt.subplots()
ax.plot(Variogram_Index,Variogram_PatternCTS[0],color='deepskyblue',linewidth=0.5,label='variogram of pattern TDS')
for index_image in range(1,RealizationAmountPerSet):
    ax.plot(Variogram_Index,Variogram_PatternCTS[index_image],color='deepskyblue',linewidth=0.5)
ax.plot(Variogram_Index,Variogram_TI,color='red',linewidth=2,label='variogram of TI')
ax.legend(loc='lower right',fontsize='xx-large')
plt.show()
```
