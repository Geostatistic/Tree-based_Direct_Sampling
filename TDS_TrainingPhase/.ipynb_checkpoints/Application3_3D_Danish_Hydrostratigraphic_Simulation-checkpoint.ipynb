```python
# import necessary package
import matplotlib.pyplot as plt
import numpy as np
import time

from sklearn.decomposition import PCA
import random
```

```python
# set the bound of unknown point
knownBelowbound = -900

# the radius of template. The actual size of template is 2*radius+1
template_radius_x = 3
template_radius_y = 8
template_radius_z = 8

# the height of the tree
TreeHeight_max = 10

# the weight of primary variable and secondary variable
Weight1 = 0.5
Weight2 = 1 - Weight1
```

```python
# set some temporary variable
point_z = 0
point_y = 0
point_x = 0

index_Pattern = 0
```

```python
# start time 0
start_time0 = time.time()
```

```python
# Read the Denmark groundwater materials

# read the three-dimensional training image
TI_Layer = 39
TI_Height = 133
TI_Width = 229
TI = np.loadtxt('Denmark_TI_conceptual_lithology_model.txt',skiprows=3).reshape(TI_Layer,TI_Height,TI_Width)

# for point_z in range(TI_Layer):
#     print("layer "+str(point_z)+" :")
#     plt.imshow(TI[point_z,:,:],vmin=-1,vmax=2,cmap='jet')
#     plt.colorbar(fraction=0.02)
#     plt.show()
    
# plt.imshow(TI[0,:,:])
# plt.colorbar(fraction=0.02)
# plt.show()

# plt.imshow(TI[:,0,:])
# plt.colorbar(fraction=0.02)
# plt.show()

# plt.imshow(TI[:,:,0])
# plt.colorbar(fraction=0.02)
# plt.show()
```

```python
# Read the Denmark groundwater materials  geophysical data

SoftData = np.loadtxt('DenmarkGroundwater_3D_SoftData_Uniform.txt',skiprows=3).reshape(TI_Layer,TI_Height,TI_Width)
# set the max difference within secondary variable
SecondaryMaxDifference = 1.00 - 0.542189    # this is the max difference between maximum and minimum

# SoftData = np.loadtxt('Denmark_skytem_soft_data.txt',skiprows=3).reshape(TI_Layer,TI_Height,TI_Width)
# # set the max difference within secondary variable
# SecondaryMaxDifference = 2.99 - (-0.11)

# for point_z in range(TI_Layer):
#     print("layer "+str(point_z)+" :")
#     plt.imshow(SoftData[point_z,:,:])
#     plt.colorbar(fraction=0.02)
#     plt.show()
```

```python
# start time 1
start_time1 = time.time()
```

```python
# extract patterns to calculate PCA eigenvector
# find the locations can be used to capture

ExtractStride_z = 4
ExtractStride_y = 4
ExtractStride_x = 4

ExtractPattern_Z = []
ExtractPattern_Y = []
ExtractPattern_X = []

PatternAmount = 0
PatternSize = (2*template_radius_z+1)*(2*template_radius_y+1)*(2*template_radius_x+1)
range_Z = ExtractStride_z * np.linspace(0,(TI_Layer-2*template_radius_z-1)//ExtractStride_z,1+((TI_Layer-2*template_radius_z-1)//ExtractStride_z)).astype(int) + template_radius_z
range_Y = ExtractStride_y * np.linspace(0,(TI_Height-2*template_radius_y-1)//ExtractStride_y,1+((TI_Height-2*template_radius_y-1)//ExtractStride_y)).astype(int) + template_radius_y
range_X = ExtractStride_x * np.linspace(0,(TI_Width-2*template_radius_x-1)//ExtractStride_x,1+((TI_Width-2*template_radius_x-1))//ExtractStride_x).astype(int) + template_radius_x
# find the location where we can extract training patterns
for point_z in range_Z:
    for point_y in range_Y:
        for point_x in range_X:
            vector = TI[(point_z-template_radius_z):(point_z+template_radius_z+1),(point_y-template_radius_y):(point_y+template_radius_y+1), (point_x-template_radius_x):(point_x+template_radius_x+1)]
            vector = vector.reshape((1,PatternSize))
            if( len(vector[ vector < knownBelowbound]) == 0 ):
                ExtractPattern_Z.append(point_z)
                ExtractPattern_Y.append(point_y)
                ExtractPattern_X.append(point_x)
                PatternAmount += 1

print('the size of pattern is '+str(PatternSize))
print('the amount of training pattern is '+str(PatternAmount))
PatternBase = np.empty((PatternAmount,PatternSize))  

# print(type(ExtractPattern_Y))
ExtractPattern_Z = np.array(ExtractPattern_Z)
ExtractPattern_Y = np.array(ExtractPattern_Y)
ExtractPattern_X = np.array(ExtractPattern_X)
# print(type(ExtractPattern_Y))
print(ExtractPattern_Z.shape)
```

```python
# extract all patterns and variable convertion
PatternBase = np.empty((PatternAmount,2*PatternSize))

# categorical variable conversion. The reason of these value can be found in Chen's Physical Review E paper
ConvertState0 = [1,0.5,0]
ConvertState1 = [0,0.5*np.sqrt(3),0]

index_Pattern = 0
for point_z, point_y, point_x in zip(ExtractPattern_Z, ExtractPattern_Y, ExtractPattern_X):
    vector = TI[(point_z-template_radius_z):(point_z+template_radius_z+1),(point_y-template_radius_y):(point_y+template_radius_y+1), (point_x-template_radius_x):(point_x+template_radius_x+1)]
    vector = vector.reshape(PatternSize)
    for index_element,element in enumerate(vector):
        state0 = ConvertState0[int(element)]
        state1 = ConvertState1[int(element)]
        PatternBase[index_Pattern][2*index_element] = state0
        PatternBase[index_Pattern][2*index_element+1] = state1
    index_Pattern += 1
# print('after convertion:'+str(PatternBase[0]))
```

```python
#perform the PCA to calculate the variance distribution and transform the previous data to new feature space
EnergyThreshold = 0.80

pca = PCA(n_components=PatternSize)
pca.fit(PatternBase)
varianceVector = pca.explained_variance_ratio_

# output the PCA variance curve and result
cumulativeVarianceVector = np.cumsum(varianceVector)
SelectDimension = np.searchsorted(cumulativeVarianceVector, EnergyThreshold)+1
SelectDimension += 1   # the last dimension is reversed for the secondary variable
print("The reduced dimension is "+str(SelectDimension))

fig, ax = plt.subplots()
ax.plot(np.linspace(1,PatternSize,PatternSize),cumulativeVarianceVector,color='r',linewidth=2,label='cumulative Variance')
ax.legend(loc='upper right')
ax.set_yticks(np.linspace(0,1.1,12))
ax.set_ylabel('cumulative variance')
ax.set_xlabel('dimension')
plt.show()
```

```python
# extract all pattern    specify the location of training patterns
ExtractStride_z = 1
ExtractStride_y = 1
ExtractStride_x = 1

ExtractPattern_Z = []
ExtractPattern_Y = []
ExtractPattern_X = []

PatternAmount = 0
PatternSize = (2*template_radius_z+1)*(2*template_radius_y+1)*(2*template_radius_x+1)
range_Z = ExtractStride_z * np.linspace(0,(TI_Layer-2*template_radius_z-1)//ExtractStride_z,1+((TI_Layer-2*template_radius_z-1)//ExtractStride_z)).astype(int) + template_radius_z
range_Y = ExtractStride_y * np.linspace(0,(TI_Height-2*template_radius_y-1)//ExtractStride_y,1+((TI_Height-2*template_radius_y-1)//ExtractStride_y)).astype(int) + template_radius_y
range_X = ExtractStride_x * np.linspace(0,(TI_Width-2*template_radius_x-1)//ExtractStride_x,1+((TI_Width-2*template_radius_x-1))//ExtractStride_x).astype(int) + template_radius_x
for point_z in range_Z:
    for point_y in range_Y:
        for point_x in range_X:
            vector = TI[(point_z-template_radius_z):(point_z+template_radius_z+1),(point_y-template_radius_y):(point_y+template_radius_y+1), (point_x-template_radius_x):(point_x+template_radius_x+1)]
            vector = vector.reshape((1,PatternSize))
            if( len(vector[ vector < knownBelowbound]) == 0 ):
                ExtractPattern_Z.append(point_z)
                ExtractPattern_Y.append(point_y)
                ExtractPattern_X.append(point_x)
                PatternAmount += 1

print('the size of pattern is '+str(PatternSize))
print('the amount of training pattern is '+str(PatternAmount))  

# print(type(ExtractPattern_Y))
ExtractPattern_Z = np.array(ExtractPattern_Z)
ExtractPattern_Y = np.array(ExtractPattern_Y)
ExtractPattern_X = np.array(ExtractPattern_X)
```

```python
# calculate the PCA score vector via batch
PatternAmount_Batch = 5000
PatternBase = np.empty((PatternAmount_Batch,2*PatternSize))

index_Pattern = 0
firstBatch = True
for point_z, point_y, point_x in zip(ExtractPattern_Z, ExtractPattern_Y, ExtractPattern_X):
    vector = TI[(point_z-template_radius_z):(point_z+template_radius_z+1),(point_y-template_radius_y):(point_y+template_radius_y+1), (point_x-template_radius_x):(point_x+template_radius_x+1)]
    vector = vector.reshape(PatternSize)
    for index_element,element in enumerate(vector):
        state0 = ConvertState0[int(element)]
        state1 = ConvertState1[int(element)]
        PatternBase[index_Pattern][2*index_element] = state0
        PatternBase[index_Pattern][2*index_element+1] = state1
    index_Pattern += 1
    
    if(index_Pattern >= PatternAmount_Batch and firstBatch == True):
        # print('shape of pattern base')
        # print(PatternBase.shape)
        PatternBase_Decompose_All = pca.transform(PatternBase)
        Pattern_Decompose_Partial_primary = np.copy(PatternBase_Decompose_All[0:PatternAmount_Batch,0:SelectDimension])
        # print('shape of PCA base')
        # print(PatternBase_Decompose_All.shape)
        # print('shape of reduced PCA base')
        # print(Pattern_Decompose_Partial_primary.shape)
        firstBatch = False
        index_Pattern = 0
    elif(index_Pattern >= PatternAmount_Batch and firstBatch == False):
        # print('shape of pattern base')
        # print(PatternBase.shape)
        PatternBase_Decompose_temporary = pca.transform(PatternBase)
        PatternBase_Decompose_temporary = np.copy(PatternBase_Decompose_temporary[0:PatternAmount_Batch,0:SelectDimension])
        # print('shape of temporary PCA base')
        # print(PatternBase_Decompose_temporary.shape)
        Pattern_Decompose_Partial_primary = np.concatenate((Pattern_Decompose_Partial_primary,PatternBase_Decompose_temporary),axis=0)
        print('shape of PCA base')
        print(Pattern_Decompose_Partial_primary.shape)
        index_Pattern = 0
    
PatternBase_Decompose_temporary = pca.transform(PatternBase[0:index_Pattern])
PatternBase_Decompose_temporary = np.copy(PatternBase_Decompose_temporary[0:PatternAmount_Batch,0:SelectDimension])
Pattern_Decompose_Partial_primary = np.concatenate((Pattern_Decompose_Partial_primary,PatternBase_Decompose_temporary),axis=0)
print('shape of PCA base')
print(Pattern_Decompose_Partial_primary.shape)

PatternSize = 2*PatternSize
```

```python
# set the benchmark state   this cell is used to convert the categorical variable
benchmark2 = np.full(shape=(1,PatternSize//2),fill_value=2)
benchmark1 = np.full(shape=(1,PatternSize//2),fill_value=1)
benchmark0 = np.full(shape=(1,PatternSize//2),fill_value=0)
benchmarkMatrix = np.stack((benchmark0,benchmark1,benchmark2),axis=0).reshape((3,PatternSize//2))
# print('benchmarkMatrix:')
# print(benchmarkMatrix)
# print('the shape of benchmark matrix is '+ str(benchmarkMatrix.shape))

# convert function from categorical to continuous
benchmarkMatrix_Continuous = []
for element in benchmarkMatrix.reshape(PatternSize*3//2):
    benchmarkMatrix_Continuous.append(ConvertState0[element])
    benchmarkMatrix_Continuous.append(ConvertState1[element])
benchmarkMatrix_Continuous = np.array(benchmarkMatrix_Continuous)
benchmarkMatrix_Continuous = benchmarkMatrix_Continuous.reshape((3,PatternSize))
# print('benchmarkMatrix after convertion:')
# print(benchmarkMatrix_Continuous)

benchmarkMatrix_PCA = pca.transform(benchmarkMatrix_Continuous)
benchmarkMatrix_PCA = benchmarkMatrix_PCA[:,0:SelectDimension]
# print('benchmarkMatrix after PCA dimension reduction:')
# print(benchmarkMatrix_PCA)
# plt.scatter(benchmarkMatrix_PCA[:,0],benchmarkMatrix_PCA[:,1], c=np.linspace(0,2,3),cmap='jet',vmin=-1,vmax=2)
# plt.colorbar()
# plt.show()

benchmark_distance01 = np.linalg.norm(benchmarkMatrix_PCA[0] - benchmarkMatrix_PCA[1])
benchmark_distance02 = np.linalg.norm(benchmarkMatrix_PCA[0] - benchmarkMatrix_PCA[2])
benchmark_distance12 = np.linalg.norm(benchmarkMatrix_PCA[2] - benchmarkMatrix_PCA[1])
benchmark_distance = (benchmark_distance01+benchmark_distance02+benchmark_distance12)/3
# print('the benchmark distance is '+str(benchmark_distance01))
# print('the benchmark distance is '+str(benchmark_distance02))
# print('the benchmark distance is '+str(benchmark_distance12))
print('the average benchmark distance is '+str(benchmark_distance))
```

```python
# multiply benchmark distance with the primary variable. Normalize the distance between primary variable pattern
print('the benchmark distance in terms of primary variable is '+str(benchmark_distance))
Pattern_Decompose_Partial_primary = Pattern_Decompose_Partial_primary/benchmark_distance
print(Pattern_Decompose_Partial_primary[0])
```

```python
# set the seconodary variable
# the size of pattern in auxiliary variable domain is 1
PatternBase_Secondary = np.empty((PatternAmount,1))
secondaryVariable = 0
index_Pattern = 0
for point_z, point_y, point_x in zip(ExtractPattern_Z, ExtractPattern_Y, ExtractPattern_X):
    secondaryVariable = SoftData[point_z,point_y,point_x]
    PatternBase_Secondary[index_Pattern] = secondaryVariable
    index_Pattern += 1

# assign the max difference weight
# print('the benchmark distance in terms of secondary variable is '+str(SecondaryMaxDifference))
PatternBase_Secondary = np.multiply(PatternBase_Secondary, 1/SecondaryMaxDifference)
```

```python
# stack primary and secondary variable
Pattern_Decompose_Partial = np.copy(Pattern_Decompose_Partial_primary)
Pattern_Decompose_Partial[:,-1] = PatternBase_Secondary.reshape(PatternAmount)
# print(Pattern_Decompose_Partial_primary.shape)
# print(Pattern_Decompose_Partial.shape)
# print(Pattern_Decompose_Partial_primary[0])
# print(Pattern_Decompose_Partial[0])
```

```python
# my own weighted distance computation
def MyWeightDistance(pattern1, pattern2,weight1,weight2):
    distance2 = np.linalg.norm(pattern1[-1]-pattern2[-1])
    distance1 = np.linalg.norm(pattern1[:-1]-pattern2[:-1])
    return weight1 * distance1 + weight2 * distance2
```

```python
# vectorize my own weighted distance computation
def MyWeightDistance_Array(pattern1, pattern2,weight1,weight2):
    instanceAmount = pattern1.shape[0]
    dimension = pattern1.shape[1]
    distance2 = np.linalg.norm(pattern1[:,-1].reshape((instanceAmount,1)) - pattern2[-1].reshape((1,1)),axis=1)
    distance1 = np.linalg.norm(pattern1[:,:-1].reshape((instanceAmount,dimension-1))-pattern2[:-1].reshape((1,dimension-1)),axis=1)
    return weight1 * distance1 + weight2 * distance2
```

```python
# alleviate memory issue
Pattern_Decompose_All = []
PatternBase_Secondary = []
PatternBase = []
```

```python
# my own binary cluster 
def MyBinaryCluster(InstanceArray,weight1,weight2):
    #initalizing cluster variable
    instanceNumber = InstanceArray.shape[0]
    # print(instanceNumber)
    cluster = np.zeros(instanceNumber)
    # print(cluster)
    
    if(instanceNumber==1):
        return cluster
    
    # center0 = np.copy(InstanceArray[0])
    center0 = np.copy(InstanceArray[np.random.randint(low=0,high=instanceNumber)])
    center1 = np.copy(InstanceArray[np.random.randint(low=0,high=instanceNumber)])
    iteration = 0
    while(MyWeightDistance(center0,center1,weight1,weight2)==0 and iteration < 100):
        center1 = np.copy(InstanceArray[np.random.randint(low=0,high=instanceNumber)])
        iteration += 1

    iteration = 0
    error = 1.0
    while(error > 0.000001 and iteration < 100):
        iteration += 1
        distanceMatrix0 = MyWeightDistance_Array(InstanceArray,center0,weight1,weight2)
        distanceMatrix1 = MyWeightDistance_Array(InstanceArray,center1,weight1,weight2)
        
        cluster = np.zeros(instanceNumber)
        cluster[ np.where(distanceMatrix0 > distanceMatrix1) ] = 1
        
        center_old0 = np.copy(center0)
        center_old1 = np.copy(center1)
        
        center0 = np.mean(InstanceArray[ np.where(cluster==0) ],axis=0)
        center1 = np.mean(InstanceArray[ np.where(cluster==1) ],axis=0)
        
        error = MyWeightDistance(center_old0,center0,weight1,weight2) + MyWeightDistance(center_old1,center1,weight1,weight2)
        
    radius0 = np.mean(MyWeightDistance_Array(InstanceArray[ np.where(cluster==0) ], center0, weight1, weight2))
    radius1 = np.mean(MyWeightDistance_Array(InstanceArray[ np.where(cluster==1) ], center1, weight1, weight2))
    return cluster, center0, center1, radius0, radius1
```

```python
# build the clustering tree
Representative_Z = []
Representative_Y = []
Representative_X = []
InnerRadius = []

treeHeight = 1

clusterAmount_old = 1
clusterIndex_old = 0
clusterAmount_new = 2

clusterResult_old = np.ones(PatternAmount).astype(int)

memberPatternIndex_old = np.linspace(0,PatternAmount-1,PatternAmount).reshape(1,PatternAmount)
memberPatternIndex_old = memberPatternIndex_old.astype(int)

KmeansResult = np.ones(PatternAmount)

MaxClusterSize = PatternAmount

RepresentativeIndexGroup = []

for treeHeight in range(1,TreeHeight_max+1):
    print('height: '+str(treeHeight))
    clusterAmount_old = 2**(treeHeight-1)
    clusterAmount_new = 2**treeHeight
    print('cluster amount:'+str(clusterAmount_new))
    
    memberPatternIndex_new = np.full(shape=(clusterAmount_new,MaxClusterSize),fill_value=-1)
    clusterSize = np.empty(clusterAmount_new)
    
    for clusterIndex_old in range(clusterAmount_old):
                
        memberPatternIndex_current = memberPatternIndex_old[clusterIndex_old]
        memberPatternIndex_current = memberPatternIndex_current[memberPatternIndex_current > -1]
        # print(memberPatternIndex_current)
        # print(len(memberPatternIndex_current))
        if(len(memberPatternIndex_current)==0):   # solve empty cluster
            for clusterIndex_child in range(0,2):
                clusterSize[clusterIndex_old*2+clusterIndex_child] = 0
                Representative_X.append( -1 )
                Representative_Y.append( -1 ) 
                Representative_Z.append( -1 )
                InnerRadius.append( -1.0 )
            continue
        elif(len(memberPatternIndex_current)==1):
            clusterSize[clusterIndex_old*2] = 1
            index_Pattern = (int)(len(Representative_X)/2 - 1)
            print('the len of current cluster is '+str(index_Pattern))
            Representative_X.append( Representative_X[index_Pattern] )
            Representative_Y.append( Representative_Y[index_Pattern] ) 
            Representative_Z.append( Representative_Z[index_Pattern] )
            InnerRadius.append( InnerRadius[index_Pattern] )
            
            clusterSize[clusterIndex_old*2+1] = 0
            Representative_X.append( -1 )
            Representative_Y.append( -1 ) 
            Representative_Z.append( -1 )
            InnerRadius.append( -1.0 )
            continue
        currentPatternBase = Pattern_Decompose_Partial[memberPatternIndex_current]
        # print('the shape of current clustering PCA matrix is '+str(currentPatternBase.shape))
        
        kmeansResult_temporary, center0, center1, radius0, radius1 = MyBinaryCluster(np.copy(currentPatternBase),Weight1,Weight2)
        
        for clusterIndex_child in range(0,2):
            # assign the index of the first cluster to the new vector
            memberPatternIndex_current = np.where(kmeansResult_temporary == clusterIndex_child)[0] # these indices are the indice within current cluster
            # print('the size of the cluster is '+str(memberPatternIndex_current.shape))
            clusterSize[clusterIndex_old*2+clusterIndex_child] = len(memberPatternIndex_current)
            memberPatternIndex_current = memberPatternIndex_old[clusterIndex_old][memberPatternIndex_current] # these indices are indices within actual patterns
            KmeansResult[memberPatternIndex_current] = clusterIndex_old*2 + clusterIndex_child
            # print(memberPatternIndex_current)
            memberPatternIndex_current_pad = np.pad(memberPatternIndex_current,pad_width=(0,MaxClusterSize-len(memberPatternIndex_current)),mode='constant',constant_values=(0,-1))
            # print(memberPatternIndex_current)
            memberPatternIndex_new[clusterIndex_old*2+clusterIndex_child] = memberPatternIndex_current_pad
            
            if(len(memberPatternIndex_current) == 0):   # solve empty cluster
                # there is no pattern in current cluster
                Representative_X.append( -1 )
                Representative_Y.append( -1 ) 
                Representative_Z.append( -1 )
                InnerRadius.append( -1.0 )
            else:
                # find the actual patterns are closest to first center 
                if(clusterIndex_child == 0):
                    representative = np.argmin( MyWeightDistance_Array(Pattern_Decompose_Partial, center0, Weight1, Weight2) )
                    InnerRadius.append( radius0 )
                else:
                    representative = np.argmin( MyWeightDistance_Array(Pattern_Decompose_Partial, center1, Weight1, Weight2) )
                    InnerRadius.append( radius1 )
                    
                Representative_X.append( ExtractPattern_X[representative] )
                Representative_Y.append( ExtractPattern_Y[representative] ) 
                Representative_Z.append( ExtractPattern_Z[representative] )
                
                RepresentativeIndexGroup.append(representative)
        
    memberPatternIndex_old = memberPatternIndex_new.copy()
    
#     # statistically analyze the size of clusters
    print('the average size: '+ str(np.mean(clusterSize)))
    MaxClusterSize = np.max(clusterSize).astype(int)
    print('the max size: '+ str(MaxClusterSize))
    print('the min size: '+ str(np.min(clusterSize)))
    
    # print the histogram of cluster size
    plt.bar(np.linspace(0,clusterAmount_new-1,clusterAmount_new), clusterSize, width = 0.8)
    plt.title('the histogram of cluster size')
    plt.show()
```

```python
# end time 0
end_time0 = time.time()
```

```python
# output the file
Representative_Z = np.array(Representative_Z).reshape(2*clusterAmount_new-2)
Representative_Y = np.array(Representative_Y).reshape(2*clusterAmount_new-2)
Representative_X = np.array(Representative_X).reshape(2*clusterAmount_new-2)
InnerRadius = np.array(InnerRadius).reshape(2*clusterAmount_new-2)
# print(Representative_Y)
file_Z = open("z_DenmarkGroundwater_3D_python_clusterTree_Representative_Z.txt","w")
file_Y = open("z_DenmarkGroundwater_3D_python_clusterTree_Representative_Y.txt","w")
file_X = open("z_DenmarkGroundwater_3D_python_clusterTree_Representative_X.txt","w")
file_Radius = open("z_DenmarkGroundwater_3D_python_cluster_AverageDistance.txt","w")
for x,y,z,radius in zip(Representative_X,Representative_Y,Representative_Z,InnerRadius):
    file_X.write(str(x)+'\n')
    file_Y.write(str(y)+'\n')
    file_Z.write(str(z)+'\n')
    file_Radius.write(str(radius)+'\n')
file_Z.close()
file_Y.close()
file_X.close()
file_Radius.close() 
```

```python
# show the clustered training image
KmeansResult_Image = np.full(shape=(TI_Layer,TI_Height,TI_Width),fill_value=-1)
# print('the size of k means result is'+str(KmeansResult.shape))
index_Pattern = 0
for point_z, point_y, point_x in zip(ExtractPattern_Z, ExtractPattern_Y, ExtractPattern_X):
    KmeansResult_Image[point_z][point_y][point_x] = KmeansResult[index_Pattern]
    index_Pattern += 1
    
# for point_z in range(TI_Layer):
#     print("layer "+str(point_z)+" :")
#     plt.imshow(KmeansResult_Image[point_z,:,:],cmap='jet',vmin=0,vmax=2**TreeHeight_max)
#     plt.colorbar(fraction=0.02)
#     plt.show()
KmeansResult_Image = KmeansResult_Image.reshape(TI_Layer*TI_Height*TI_Width)
file_KmeansResult = open("z_DenmarkGroundwater_3D_python_clusterTree_Result.txt","w")
for cluster in KmeansResult_Image:
    file_KmeansResult.write(str(cluster)+'\n')
file_KmeansResult.close()
```
